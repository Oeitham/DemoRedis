```python
import redis
monRedis = redis.StrictRedis(decode_responses=True)

# PING message
monRedis.ping()
```

```python
# FLUSHALL
monRedis.flushall()

# KEYS pattern
monRedis.keys()
```

```python
# SET key value
monRedis.set("maCle", 1)

# GET key
monRedis.get("maCle")
```

```python
# MSET key value [key value ...]
monRedis.mset({"maCle": 3, "maCle2": 2})

# MGET key [key ...]
monRedis.mget(["maCle", "maCle2"])
```

```python
monRedis.set("maCle", "modif-nx", nx=True)
monRedis.set("maCle2", "modif-xx", xx=True)
monRedis.set("maCle3", "new-nx", nx=True)
monRedis.set("maCle4", "new-xx", xx=True)

monRedis.mget(["maCle", "maCle2", "maCle3", "maCle4"])
```

```python
# DEL key [key ...]
monRedis.delete("maCle3")

monRedis.keys()
```

```python
import random

colors = ["", "Blue", "Red", "Green", "Yellow", "Black", "White", "Pink", "Purple", "Cyan", "Gray", "Brown"]
name = ["Man", "Guy", "Dude", "Girl", "Gurl", "God", "Troll", "Princess", "Clown", "Acrobat", "Ninja", "Hero"]
adjectives = ["", "Big", "Cute", "Cynical", "Strong", "Killer", "Sniping", "Happy", "Lame", "Great", "Formidable"]
numbers = ["0", "9", "8", "7", "6", "5", "4", "3", "2", "1", "", "", "", "", "", "", "", "", "", ""]
def get_user():
    return random.choice(colors)+random.choice(adjectives)+random.choice(name)+random.choice(numbers)+random.choice(numbers)
```

```python
def add_score(user, score):
    old_score = monRedis.zscore("PLayerScores", user)
    if not old_score or score > int(old_score):
        # ZADD key
        monRedis.zadd("PlayerScores", {user: score})

for i in range(100000):
    user = get_user()
    score = int(random.normalvariate(50000, 25000))
    add_score(user, score)
```

```python
# ZCOUNT key min max
monRedis.zcount("PlayerScores", float('-inf'), float('inf'))
```

```python
# ZRANGE key min max
monRedis.zrange("PlayerScores", 0, 9, withscores=True, desc=True)
```

```python
add_score("BlueKillerTroll", 50000)

# ZSCORE key member
monRedis.zscore("PlayerScores", "BlueKillerTroll")
```

```python
# ZRANK key member
monRedis.zrank("PlayerScores", "BlueKillerTroll")
```

```python
nbplayer = monRedis.zcount("PlayerScores", float('-inf'), float('inf'))
rank = nbplayer - 1 - monRedis.zrank("PlayerScores", "BlueKillerTroll")
monRedis.zrange("PlayerScores", rank-4, rank+5, withscores=True, desc=True)
```

```python
for color in colors:
    # LPUSH key
    monRedis.lpush("Friends:BlueKillerTroll", color + "KillerTroll")

# LRANGE key start stop
monRedis.lrange("Friends:BlueKillerTroll", 0, 99)
```

```python
friends = monRedis.lrange("Friends:BlueKillerTroll", 0, 99)

for friend in friends:
    score = monRedis.zscore("PlayerScores", friend)
    monRedis.zadd("FriendScores:BlueKillerTroll", {friend: score})

monRedis.zrange("FriendScores:BlueKillerTroll", 0, 99, withscores=True, desc=True)
```

```python
# EXPIRE key seconds
monRedis.expire("FriendScores:BlueKillerTroll", 60)

# TTL key
monRedis.ttl("FriendScores:BlueKillerTroll")
```

```python
# HSET key field value
monRedis.hset("Players:00000001", "username", "BlueKillerTroll")
monRedis.hset("Players:00000001", "key:score", "BlueKillerTroll")
monRedis.hset("Players:00000001", "key:friends", "Friends:BlueKillerTroll")

# HGET key field
monRedis.hgetall("Players:00000001")
```
